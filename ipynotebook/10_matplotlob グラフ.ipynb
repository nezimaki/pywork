```python
def func1():
    # y-1=2x
    X=np.arange(-1.0,1.0,0.01);

    Y=2*X
    plt.plot(X,Y,color='blue'); # y=2x

    Y=2*X+1
    plt.plot(X,Y,color='red'); 

    plt.plot([-1.0,1.0],[0.0,0.0],color='black') 
    # x 軸 
    plt.plot([0.0,0.0],[-2.0,3.0],color='black') # y 軸 
    plt.axis([-1.1,1.1,-2.1,3.1])

    plt.title(r'$y=2x,\ y-1=2x$',fontsize=18); 
    plt.xlabel(r'$x$',fontsize=18); 
    plt.ylabel(r'$y$',fontsize=18); 
    plt.legend((r'$y=2x$',r'$y-1=2x$'),loc='lower right') 
    plt.grid(True); # グラフにグリッドを付ける
    plt.show();
    
func1()
```

```python
"""
カラーマップを利用する
"""
import matplotlib.cm as cm

X=np.arange(0.0,1.0,0.01)

for n in range(1,7):
    Y=np.sqrt(X*(1-X) / n)
    plt.plot(X,Y,color=cm.gist_rainbow(n/7.0), label='n=%s' % n)
    
plt.legend()
```

```python
"""
カラーマップを利用する(2)
"""

import matplotlib.cm as cm

X=np.arange(0.0,100.0,1)

for n in range(1,10):
    Y= n / np.sqrt(X)
    plt.plot(X,Y,color=cm.gist_rainbow(n/7.0), label='d=%s' % n)
    
plt.legend()
plt.xlabel(r'$x$',fontsize=18); 
plt.ylabel(r'$y$',fontsize=18); 
plt.grid(True);
plt.show();
```

```python
#!/usr/bin/env python
#-*- coding:utf-8 -*-

"""
numpyのfftpackを利用したフーリエ変換
"""

import numpy as np
import matplotlib.pyplot as pl

#サンプリングレート
sample_rate = 44100.0

#サンプル点の準備
t= np.r_[0:0.1:1/sample_rate]

#サンプル波
y = np.sin((2*np.pi)*30*t)+2*np.cos((2*np.pi)*80*t)

#サンプル波を表示
pl.plot(t,y,"y")
pl.show()

#fftpackでfft
f = np.fft.fft(y)

#fftpackから周波数ビンの作成
freq = np.fft.fftfreq(t.shape[-1],1/sample_rate)

pl.plot(freq[:30],f.real[:30],"r")
pl.show()

pl.plot(freq[:30],f.imag[:30],"b")
pl.show()

pl.plot(freq[:30],np.abs(f[:30]))
pl.show()
```
